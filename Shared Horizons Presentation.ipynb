```python
import meter_graph

mg = MeterGraph()

def translate(s):
    return mg.graph_scan(s)
```

```python
#s=' ranj se ;xuugar hu))aa insaa;n to mi;t jaataa hai ranj'
s=' ja;zbah-e be-i;xtiyaar-e shauq dekhaa chaahiye'
translations = translate(s)
for t in translations:
    print "scan: ",t.scan, "meter id: ",t.meter_type
    print "Also saves matches, e.g.:"
    pprint(t.matches[0:5], indent=4)
    print t.meter_type
```
